```python
diamonds = diamonds[diamonds.carat<3]
diamonds = diamonds[diamonds.depth<72]
diamonds = diamonds[diamonds.depth<70]
```

## train

```python
X = diamonds.drop(['price'], axis=1)
y = diamonds['price']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=69)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
p_test = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features':["auto",'sqrt','log2'],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
```

```python
tuning=GridSearchCV(estimator=RandomForestRegressor(), param_grid=p_test, n_jobs = -1, cv = 3, verbose =2)
tuning.fit(X_train,y_train)
tuning.score(X_test, y_test)
pred= tuning.predict(X_test)
r2 = r2_score(y_test, pred)
r2
```
